```python
# Question 3
# Find number of bills greater than daily average and display them (tips data)
df=pd.read_csv('tips.csv')
pd.set_option('display.max_rows', None)
df.groupby('day')[['total_bill']].mean()  
```

```python
# Question 6
# Read this excel file and bring only the records containing at least one null value (tips data)
df=pd.read_csv('TIPS_data2.csv')
df.isna().any()
df[df.isnull().any(axis=1)]
```

```python
# Question 7
# How many people have the same name and surname? (tips data)
df=pd.read_csv('TIPS_data2.csv')
df[df['Name']==df['Surname']]

df_new=df.groupby(['Name','Surname']).size().reset_index(name='count')
df_new[df_new['count']>1]
```
